```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lectura
from google.colab import drive
drive.mount('/content/drive')
ruta = "/content/drive/MyDrive/Colab Notebooks/Data/dataset_banco.csv"
data_frame = pd.read_csv(ruta)
```

```python
print(data_frame.shape)
data_frame.head(20)
```

```python
data_frame_rename = data_frame.rename(columns={"age" : "Edad",
                                               "job" : "Tipo de Trabajo",
                                               "marital" : "Estado Civil",
                                               "education" : "Nivel Educativo",
                                               "default" : "Morosidad",
                                               "balance" : "Saldo",
                                               "housing" : "Vivienda",
                                               "loan" : "Prestamo",
                                               "contact" : "Contacto",
                                               "day" : "Día",
                                               "month" : "Mes",
                                               "duration" : "Duración",
                                               "campaign" : "Campaña",
                                               "pdays" : "Días Previos",
                                               "previous": "Anterior",
                                               "poutcome" : "Resultados Previos",
                                               "y" : "Respuesta" })
print(data_frame_rename)
```

```python
data_frame_rename.info()
```

```python
# Clasificar columnas
categoricas = data_frame_rename.select_dtypes(include=['object']).columns.tolist()
numericas = data_frame_rename.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Columnas categóricas:", categoricas)
print("Columnas numéricas:", numericas)
```

```python
data_frame_2 = data_frame_rename
print(data_frame_2)
```

```python
data_frame_2.describe()
```

```python
#Eliminar valores data frame
data_frame_2.dropna(inplace=True)
data_frame_2.count()
```

```python

#Conteo de los niveles en las diferentes columnas categóricas
cols_cat = ["Tipo de Trabajo", "Estado Civil", "Nivel Educativo", "Morosidad", "Vivienda", "Prestamo", "Contacto", "Mes", "Resultados Previos", "Respuesta"]
#print(cols_cat)
for col in cols_cat:
  print(f"Columna {col} : {data_frame_2[col].nunique()} subniveles")
```

```python
pd.value_counts(data_frame_2["Tipo de Trabajo"])
```

```python
#Calcular desviación estandar de columnas numericas
desviacion_estandar = data_frame_2[['Edad', 'Saldo', 'Día', 'Duración', 'Campaña', 'Días Previos', 'Anterior']].std()
print(desviacion_estandar)
```

```python
destd_dias_previos = data_frame_2[["Días Previos"]].std()
print(destd_dias_previos)
```

```python
desviacion_estandar.plot(kind ="bar")
plt.title("Desviación Estándar de las Variables Numéricas")
plt.xlabel("Variables")
plt.ylabel("Desviación Estándar")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Graficar la desviación estándar de días previos
plt.figure(figsize=(4,2))
plt.bar("Días Previos", destd_dias_previos)
plt.title("Desviación Estándar de Días Previos")
plt.ylabel("Desviación Estándar")
plt.show()
```

```python
#Calcula la media del atributo "Días Previos: dias de contacto al usuario antes de la campaña actual"
media = data_frame_2["Días Previos"].mean()
print("Promedio de días de contacto", media)
```

```python
#Graficar un histograma del atributo "Días Previos"
plt.hist(data_frame_2['Días Previos'], bins=20, color='skyblue', edgecolor='black')

#Agregar una línea vertical para mostrar la media
plt.axvline(media, color = "red", linestyle = "dashed", linewidth = 1)
# Etiquetas y título
plt.xlabel('Valor de Días Previos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Días Previos con Media')
#Mostrar grafica
plt.show()
```

```python
#Generar graficas para cada una de las variables numericas

cols_num = ['Edad', 'Saldo', 'Día', 'Duración', 'Campaña', 'Días Previos', 'Anterior']
fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)
for i, col in enumerate(cols_num):
    sns.boxplot(x=col, data=data_frame_2, ax=ax[i])
    ax[i].set_title(col)
```
