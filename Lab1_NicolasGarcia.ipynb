```python
# Importar librer√≠as
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lectura
from google.colab import drive
drive.mount('/content/drive')
ruta = "/content/drive/MyDrive/Colab Notebooks/Data/dataset_banco.csv"
data_frame = pd.read_csv(ruta)
```

```python
print(data_frame.shape)
data_frame.head(20)
```

```python
data_frame_rename = data_frame.rename(columns={"age" : "Edad",
                                               "job" : "Tipo de Trabajo",
                                               "marital" : "Estado Civil",
                                               "education" : "Nivel Educativo",
                                               "default" : "Morosidad",
                                               "balance" : "Saldo",
                                               "housing" : "Vivienda",
                                               "loan" : "Prestamo",
                                               "contact" : "Contacto",
                                               "day" : "D√≠a",
                                               "month" : "Mes",
                                               "duration" : "Duraci√≥n",
                                               "campaign" : "Campa√±a",
                                               "pdays" : "D√≠as Previos",
                                               "previous": "Anterior",
                                               "poutcome" : "Resultados Previos",
                                               "y" : "Respuesta" })
print(data_frame_rename)
```

```python
data_frame_rename.info()
```

```python
# Clasificar columnas
categoricas = data_frame_rename.select_dtypes(include=['object']).columns.tolist()
numericas = data_frame_rename.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Columnas categ√≥ricas:", categoricas)
print("Columnas num√©ricas:", numericas)
```

```python
pd.value_counts(data_frame_rename['Tipo de Trabajo'])
```

```python
data_frame_rename.isna().sum()
```

```python
data_frame_2 = data_frame_rename
print(data_frame_2)
```

```python
data_frame_2.describe()
```

```python
#Eliminar valores data frame
data_frame_2.dropna(inplace=True)
data_frame_2.count()
```

```python

#Conteo de los niveles en las diferentes columnas categ√≥ricas
cols_cat = ["Tipo de Trabajo", "Estado Civil", "Nivel Educativo", "Morosidad", "Vivienda", "Prestamo", "Contacto", "Mes", "Resultados Previos", "Respuesta"]
#print(cols_cat)
for col in cols_cat:
  print(f"Columna {col} : {data_frame_2[col].nunique()} subniveles")
```

```python
pd.value_counts(data_frame_2["Tipo de Trabajo"])
```

```python
#Calcular desviaci√≥n estandar de columnas numericas
desviacion_estandar = data_frame_2[['Edad', 'Saldo', 'D√≠a', 'Duraci√≥n', 'Campa√±a', 'D√≠as Previos', 'Anterior']].std()
print(desviacion_estandar)
```

```python
destd_dias_previos = data_frame_2[["D√≠as Previos"]].std()
print(destd_dias_previos)
```

```python
desviacion_estandar.plot(kind ="bar")
plt.title("Desviaci√≥n Est√°ndar de las Variables Num√©ricas")
plt.xlabel("Variables")
plt.ylabel("Desviaci√≥n Est√°ndar")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Graficar la desviaci√≥n est√°ndar de d√≠as previos
plt.figure(figsize=(4,2))
plt.bar("D√≠as Previos", destd_dias_previos)
plt.title("Desviaci√≥n Est√°ndar de D√≠as Previos")
plt.ylabel("Desviaci√≥n Est√°ndar")
plt.show()
```

```python
#Calcula la media del atributo "D√≠as Previos: dias de contacto al usuario antes de la campa√±a actual"
media = data_frame_2["D√≠as Previos"].mean()
print("Promedio de d√≠as de contacto", media)
```

```python
#Graficar un histograma del atributo "D√≠as Previos"
plt.hist(data_frame_2['D√≠as Previos'], bins=20, color='skyblue', edgecolor='black')

#Agregar una l√≠nea vertical para mostrar la media
plt.axvline(media, color = "red", linestyle = "dashed", linewidth = 1)
# Etiquetas y t√≠tulo
plt.xlabel('Valor de D√≠as Previos')
plt.ylabel('Frecuencia')
plt.title('Histograma de D√≠as Previos con Media')
#Mostrar grafica
plt.show()
```

```python
#Generar graficas para cada una de las variables numericas

cols_num = ['Edad', 'Saldo', 'D√≠a', 'Duraci√≥n', 'Campa√±a', 'D√≠as Previos', 'Anterior']
fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)
for i, col in enumerate(cols_num):
    sns.boxplot(x=col, data=data_frame_2, ax=ax[i])
    ax[i].set_title(col)
```
